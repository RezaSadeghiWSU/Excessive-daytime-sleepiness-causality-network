```python
# Checks the column called "modified_complete" for NaNs 
# If empty, then it checks how much of the row has been filled out and if it doesn't pass the threshold, removes it

def check_completion(df, threshold):

    length_rows = df.shape[1]
    
    #Iterate through each row of the dataframe
    for index, row in df.iterrows():
        
        # If there is a NaN under the "modified_complete" column:
        if pd.isna(row['modified_completed']):
             
            # Check what percentage of data is available in that row
            percentage_nans = row.isna().sum() / length_rows * 100 

            # Drop the row if too many NaNs
            if percentage_nans >= threshold:
                modified_df = df.drop(index) #the drop function does not modify the original dataframe unless inplace=True
                print(f'Row {index} has {round(percentage_nans,2)}% NaNs. Row {index} will be removed from dataframe.')
    
    return modified_df
    
```
